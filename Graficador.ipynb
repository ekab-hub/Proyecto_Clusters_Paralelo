# Importar

```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

ruta_archivo = 'datos_clusters.csv' 
df = pd.read_csv(r'C:\Users\HP15\Proyecto_Clusters_Paralelo\datos_2d_100000_clusters.csv')
```

## Grafica 3d

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

colors = ["red", "blue", "orange", "black", "purple", "pink", "brown", "gray"]

clusters_identificados = sorted(df['cluster'].unique())

for i, cluster_id in enumerate(clusters_identificados):
    cluster_points = df[df["cluster"] == cluster_id]
    
    ax.scatter(
        cluster_points["x"],
        cluster_points["y"],
        cluster_points["z"],
        marker="o",
        s=10,                      
        color=colors[i % len(colors)],
        alpha=0.3,                 
        label=f'Cluster {cluster_id}'
    )

ax.set_title("Cluster 3D", fontsize=18)
ax.set_xlim(-0.1, 1.1)
ax.set_ylim(-0.1, 1.1)
ax.set_zlim(-0.1, 1.1)

ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

plt.show()
```
