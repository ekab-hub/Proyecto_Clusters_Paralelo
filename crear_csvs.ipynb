```python
import numpy as np
from sklearn.datasets import make_blobs
import os
```

```python
# Configuración
puntos = [100000, 200000, 300000, 400000, 600000, 800000, 1000000]
K        = 5
SEED     = 7

# Crear carpeta si no existe
os.makedirs("csv", exist_ok=True)

for n in puntos:
    for dims in [2, 3]:

        points, _ = make_blobs(
            n_samples=n,
            centers=K,
            n_features=dims,
            cluster_std=0.04,
            random_state=SEED,
            center_box=(0, 1.0)
        )

        points = np.round(np.abs(points), 3)

        nombre = f"csv/{n}_{dims}d.csv"
        np.savetxt(nombre, points, delimiter=",", fmt="%.3f")
        print(f"Generado: {nombre}  ({n:,} puntos, {dims}D)")
```
